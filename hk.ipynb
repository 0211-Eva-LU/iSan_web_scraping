```python
# 擷取指定區域螢幕截圖 (x, y, width, height)
pyautogui.screenshot('test_screenshot.png', region=(300, 400, 100, 100))  

# 在螢幕上尋找圖片並點擊
location = pyautogui.locateCenterOnScreen(f'test_screenshot.png', confidence=0.9)
if location:
    print(location)
    pyautogui.click(location)
```

```python
import pyautogui
import pyscreeze

pyautogui.screenshot('test_screen.png',region=(1200,700,38,45))
#需要擷取數字的大小
num_size = 30

#數字之間的水平與垂直句麗
x_distance = 80
y_distance = 50

#起始位置(從數字1位置開始)
start_location = (30,410)

for i in range(3):

    pyautogui.screenshot(f"{1+i}.png",region=(start_location[0]+x_distance*i,start_location[1],num_size,num_size))
    pyautogui.screenshot(f"{4+i}.png",region=(start_location[0]+x_distance*i,start_location[1]-y_distance,num_size,num_size))
    pyautogui.screenshot(f"{7+i}.png",region=(start_location[0]+x_distance*i,start_location[1]-y_distance*2,num_size,num_size))



pyautogui.screenshot(f"0.png",region=(start_location[0]+x_distance*1,start_location[1]+y_distance,num_size,num_size))
    


```

```python
def click_num(num):
    location = pyautogui.locateCenterOnScreen(f'{num}.png')
    if location:
        pyautogui.click(location)
```
