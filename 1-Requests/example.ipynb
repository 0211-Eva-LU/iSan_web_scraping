```python
import requests
from bs4 import BeautifulSoup as bs
url = 'https://www.ptt.cc/bbs/Gossiping/index.html'
cookies = {
    "over18" : "1"
}

r= requests.get(url,cookies=cookies)
soup = bs(r.text,'lxml')
prefix="https://www.ptt.cc/"

soup.find_all("div",class_="r-ent")
title_containers=soup.select(".r-ent .title a") #空格為下一層
for container in title_containers :
    print(f"標題{container.get_text()} 網址:{prefix+container['href']}")

elements = soup.find_all(string='‹ 上頁')
print(elements[0].parent['href'])

```

```python
url = "https://tw.stock.yahoo.com/quote/0050.TW/technical-analysis"
r = requests.get(url)
soup = bs(r.text,'lxml')
price = soup.find_all(class_="Fz(32px) Fw(b) Lh(1) Mend(16px) D(f) Ai(c) C($c-trend-down)")
name = soup.find_all(class_="D(f) Ai(c) Mb(6px)")
rname = soup.find_all(class_="C($c-link-text) Fw(b) Fz(24px) Mend(8px) Whs(nw)")
trend = soup.find_all(class_="Fz(20px) Fw(b) Lh(1.2) Mend(4px) D(f) Ai(c) C($c-trend-down)")
l_h = soup.find_all(class_='Mend(4px) Bds(s)')
color = l_h[0].get('style').split('border-color:')[1].split()[0]

# print(trend[0].get_text())
def colortrend() :
    if color == "#00ab5e" :
        return "-"
    else:
        return "+"

print(f'股價:{price[0].get_text()},股票名稱:{rname[0].get_text()},漲跌:{colortrend()}{trend[0].get_text()}')
```

```python
def get_stock_info(code):
    url = f"https://tw.stock.yahoo.com/quote/{code}.TW"
    r = requests.get(url)
    soup = bs(r.text,'lxml')

    up_down= {
    "border-color:#00ab5e transparent transparent transparent;border-width:9px 6.5px 0 6.5px":"-",
    "border-color:transparent transparent #ff333a transparent;border-width:0 6.5px 9px 6.5px":"+"
    }

    title_elemnet = soup.find("h1",class_="C($c-link-text) Fw(b) Fz(24px) Mend(8px) Whs(nw)")
    title=title_elemnet.text

    price_container = soup.find('div',class_="D(f) Ai(fe) Mb(4px)")
    price = price_container.select_one(".Fz\\(32px\\)").text
    up_or_down = soup.find('span',class_="Mend(4px) Bds(s)")['style']
    up_d = up_down[up_or_down]
    trend_element = soup.select_one(".Fz\\(20px\\).Fw\\(b\\).Lh\\(1\\.2\\).Mend\\(4px\\)")
    trend = trend_element.text
    
    print(f"股票名稱:{title}\n股票價格:{price}\n股票漲幅:{up_d}{trend}\n")

get_stock_info(4142)
get_stock_info(2330)
```
