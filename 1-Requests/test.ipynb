```python
import requests
from bs4 import BeautifulSoup as bs
url =f'https://store.line.me/stickershop/product/23870/zh-Hant'
r= requests.get(url)
soup = bs(r.text,'lxml')

```

```python
import os
import json

#建立資料夾
folder_name = "line_sticlers"
if not os.path.exists(folder_name):
    os.mkdir('line_sticlers')
else:
    print("已建立過資料夾")



#找到貼圖清單
sticker = soup.find('ul',class_="mdCMN09Ul FnStickerList")
#找到所有貼圖元素
sticker_elements = sticker.find_all("li")



for index,sticker_element in enumerate(sticker_elements):
    # print(sticker_element)
    # print(sticker_element["data-preview"])
    data_preview = json.loads(sticker_element["data-preview"])
    static_url = data_preview["staticUrl"]
    print(static_url)

    img_content = requests.get(static_url).content
    with open(f"line_sticlers/line_sticker_{index}.png","wb") as f:
        f.write(img_content)
    



```

```python
'''
課堂練習
自行選擇三種貼圖進行下載
在儲存時要自動產生該貼圖名稱的資料夾並將圖片儲存在內
'''

def linestatic(number):

    import requests
    from bs4 import BeautifulSoup as bs
    import os

    url =f'https://store.line.me/stickershop/product/{number}/zh-Hant'
    r= requests.get(url)
    soup = bs(r.text,'lxml')


    sticker_name = soup.find("p",class_="mdCMN38Item01Ttl").text

    #建立資料夾
    # folder_name = sticker_name
    if not os.path.exists(sticker_name):
        os.mkdir(sticker_name)
    else:
        print("已建立過資料夾")



    #找到貼圖清單
    sticker = soup.find('ul',class_="mdCMN09Ul FnStickerList")
    #找到所有貼圖元素
    sticker_elements = sticker.find_all("li")



    for index,sticker_element in enumerate(sticker_elements):
        # print(sticker_element)
        # print(sticker_element["data-preview"])
        data_preview = json.loads(sticker_element["data-preview"])
        static_url = data_preview["staticUrl"]
        print(static_url)

        img_content = requests.get(static_url).content
        with open(f"{sticker_name}/line_sticker{sticker_name}_{index}.png","wb") as f:
            f.write(img_content)

linestatic(27469)
```
